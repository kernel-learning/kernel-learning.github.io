```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_test, y_test = np.load('data/test_data.npy')
x_train, y_train = np.load('data/train_data.npy')

# === Plot === #
plt.figure(figsize=(5, 3))
plt.scatter(x_train, y_train, color='red', label="Train samples")
plt.scatter(x_test, y_test, color='blue', label="Test samples")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.savefig('data.pdf')
plt.show()

```
