```python
def relu(x):
    return np.maximum(0, x)

def loss(preds, y, n):
    return np.mean((y - preds)**2) / 2

def generate_relu_function(x, u, V):
    x = np.atleast_1d(x)
    W = V[:, 0][:, None]  # (m, 1)
    b = V[:, 1][:, None]  # (m, 1)
    h = relu(W @ x[None, :] + b)  
    return u @ h  # (n,)

def generate_relu_samples(n, x_range, u, V):
    x_samples = np.random.uniform(x_range[0], x_range[1], n_samples)
    y_samples = generate_relu_function(x_samples, u, V)
    return x_samples, y_samples

def test_loss(u_pred, V_pred):
    y_pred = generate_relu_function(x_dense, u_pred, V_pred)
    return np.mean((y_dense - y_pred)**2) / 2

def GD(gamma, u0, V0, num_iter, wd=0):
    losses, test_losses, us, Vs, norms = [], [], [], [], []
    grads, indic_mats, VXs = [], [], []
    u, V = u0.copy(), V0.copy()
    for i in range(1, num_iter + 1):
        VX = V @ X.T
        indic_mat = (VX >= 0)
        preds = u.T @ (VX * indic_mat)

        mat = ((preds - y) * X.T).T
        Xrts = mat.T @ indic_mat.T

        grad_V = (u * Xrts).T / n_samples + wd * V
        grad_u = np.diag(V @ Xrts) / n_samples + wd * u

        if i in xs or i == 0:
            losses.append(loss(preds, y, n_samples))
            test_losses.append(test_loss(u, V))
            us.append(u.copy())
            Vs.append(V.copy())
            indic_mats.append(indic_mat)
            VXs.append(VX)
            norms.append(np.linalg.norm(V, 'fro')**2 + np.linalg.norm(u)**2)

        grads.append(Xrts.T / n_samples)

        u -= gamma * grad_u
        V -= gamma * grad_V

    return losses, test_losses, np.array(us), np.array(Vs), np.array(norms)

def grad(u, V, X, y, n_samples):
    VX = V @ X.T
    indic_mat = (VX >= 0)
    preds = u.T @ (VX * indic_mat)

    mat = ((preds - y) * X.T).T
    Xrts = mat.T @ indic_mat.T
    return Xrts.T / n_samples
```

```python
# === Parameters === #
n_samples = 10
m_neurons = 3
x_range = (-2, 2)
n_test = 100

# Ground truth network (used to generate data)
u_true = np.array([ 1 , 1, -8])
V_true = np.array([[-2, 2], [4, 1], [1, -1]])

# data used in paper
data = np.load('data/relu_data.npz')

X = data['X']
y = data['y']
x_samples = X[:, 0]

# Dense plot grid
x_dense = np.linspace(x_range[0], x_range[1], 200)
y_dense = generate_relu_function(x_dense, u_true, V_true)

# === Plot === #
plt.figure(figsize=(8, 5))
plt.plot(x_dense, y_dense, label="Ground truth ReLU function", color='blue')
plt.scatter(x_samples, y, color='red', label="Training samples", alpha=0.7)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2-layer ReLU Network Output and Samples")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
power = 7
num_iter = int(10**power)
xs = np.unique(np.round(np.logspace(0, power, 101)))[:-1]
```

```python
# Network and training parameters
m = 100  # number of hidden units
d = 2    # input dimension
alpha = 2 # scale of init
gamma = 1 / 300 # stepsize

# Initialise parameters
np.random.seed(40)
V0 = alpha * np.random.randn(m, d)
u0 = alpha * np.random.randn(m)

# Run gradient descent with weight decay
losses_wd, test_losses_wd, us_wd, Vs_wd, norms_wd = GD(gamma, u0, V0, num_iter, wd=0.001)
preds_over_time = [generate_relu_function(x_dense, u, V) for u, V in zip(us_wd, Vs_wd)]
```

```python
stop_time = -1   # moment where you want to plot the prediction
```

```python
fig, ax1 = plt.subplots(figsize=(4.4, 3))

# Left y-axis: Losses
ax1.semilogx(xs*gamma, losses_wd, color='b', linewidth=3, label='Train')
ax1.semilogx(xs*gamma, test_losses_wd, color='b', linewidth=3, label='Test', linestyle = '--')
ax1.set_ylabel("Loss", color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_ylim(bottom = -0.05, top=2)

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()

# Right y-axis: Norms
ax2.semilogx(xs*gamma, norms_wd, color='green', linestyle='--', linewidth=3, label=r'$\Vert w_t \Vert$')
ax2.set_ylabel(r'$\Vert w_t \Vert$', color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Combine legends from both axes
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(4, 3))
plt.plot(x_dense, y_dense, label='Teacher function', color='deepskyblue', linestyle='--', linewidth = '2')
plt.plot(x_dense, preds_over_time[stop_time], label='Learnt function $f_{w_t}$', linewidth= 3, color='b')
plt.scatter(x_samples, y, marker = 'x', color = 'k', s = 100, zorder=10)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.tight_layout()
plt.show()
```
